```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Auto Sales data.csv')

df.head()
```

```python
df['SALES'].isnull().sum()
```

```python
df['SALES'].isnull().sum()
```

```python

df['SALES'].hist(bins=30, figsize=(12,8), color='skyblue', edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('count')
plt.show()
```

```python
df['MSRP'].hist(bins=30, figsize=(12,8), color='green', edgecolor='black')
plt.title('MSRP Distribution')
plt.xlabel('MSRP')
plt.ylabel('Count')
plt.show()
```
